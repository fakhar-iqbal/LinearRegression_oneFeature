```python
#importing the required libraries for this model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

```

```python
#importing data frame into the notebook
df = pd.read_csv('data.csv')
df
```

```python
#getting scatter plot of the data we have
%matplotlib inline
plt.figure(figsize=(5,3),dpi=130)
plt.xlabel('Price in US $')
plt.ylabel('Area in sq. feet')
plt.title('graph for prices in the town')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python

#getting object for linear regression
reg = linear_model.LinearRegression()
#fitting the data to linear regression 

reg.fit(df[['area']],df.price)
```

```python
#getting the prediction of our desired area. the data must be in 2d shape
reg.predict([[5000]])
```

```python
#what parameters did this model use

#w was 
print('w=', reg.coef_)
#b was
print('b = ', reg.intercept_)
```

```python
## now we are plotting the model according to all the areas and their respective predictions

%matplotlib inline
plt.figure(figsize=(5,3),dpi=130)
plt.xlabel('Price in US $')
plt.ylabel('Area in sq. feet')
plt.title('graph for prices in the town')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```

```python
#predicting prices for the file predict.csv 
pr = pd.read_csv('predict.csv')

p=reg.predict(pr)
pr['prices'] = p
pr.to_csv('predict.csv',index=False)
```
